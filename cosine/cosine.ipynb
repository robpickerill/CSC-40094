```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
u = np.array([1, 1])
v = np.array([2, 1])

# Calculate the cosine similarity
cosine_similarity = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
cosine_angle = np.arccos(cosine_similarity)

print(f"The cosine similarity between u and v is: {cosine_similarity:.2f}")
print(f"The cosine angle between u and v is: {np.degrees(cosine_angle):.2f} degrees")

# Plot the vectors
fig, ax = plt.subplots(figsize=(6, 6))
ax.quiver([0, 0], [0, 0], [u[0], v[0]], [u[1], v[1]], scale=1, units='width')
ax.set_xlim([-0.5, 2.5])
ax.set_ylim([-0.5, 2.5])
ax.set_aspect('equal')
ax.set_title(f"Cosine Angle: {np.degrees(cosine_angle):.2f} degrees")
plt.show()
```
